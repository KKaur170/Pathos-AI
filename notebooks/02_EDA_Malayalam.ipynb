```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your final merged dataset
df = pd.read_csv('malayalam_final_with_emotions.csv')  # Change if already in memory

# 1. Basic info
print("Shape:", df.shape)
print("\nColumns:\n", df.columns)
print("\nSample data:\n", df.head())

# 2. Missing values
print("\nMissing values:\n", df.isnull().sum())

# 3. Emotion distribution
emotion_counts = df['emotion'].value_counts()
print("\nEmotion distribution:\n", emotion_counts)

# 4. Bar plot of emotions
plt.figure(figsize=(8, 5))
sns.countplot(x='emotion', data=df, palette='Set2')
plt.title('Emotion Label Distribution')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 5. Correlation heatmap (audio features only)
audio_features = df.filter(regex='^mel_')  # Select mel features only
plt.figure(figsize=(10, 8))
sns.heatmap(audio_features.corr(), cmap='coolwarm', center=0)
plt.title('Correlation between Audio Features')
plt.show()
```

```python
from google.colab import files
uploaded = files.upload()

# 📦 Install required libraries (if needed)
!pip install -q imbalanced-learn

# 🧾 1. Import Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, RobustScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.feature_selection import SelectKBest, f_classif
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt

# 📁 2. Load Dataset
df = pd.read_csv("malayalam_final_with_emotions.csv")  # Replace with your actual file name

# 🧹 3. Separate features & target
target = 'emotion'
text_cols = ['text_0', 'text_1', 'text_2', 'text_3', 'text_4']
mel_cols = ['mel_124', 'mel_125', 'mel_126', 'mel_127', 'mel_128']

X = df[text_cols + mel_cols]
y = df[target]

# 🔁 4. Preprocessing
# Scale text features (Standard)
X[text_cols] = StandardScaler().fit_transform(X[text_cols])

# Log transform + robust scale mel features
X[mel_cols] = np.log1p(X[mel_cols])
X[mel_cols] = RobustScaler().fit_transform(X[mel_cols])

# ⚖ 5. Balance Classes using SMOTE
sm = SMOTE(random_state=42)
X_resampled, y_resampled = sm.fit_resample(X, y)

# 🎯 6. Feature Selection (Optional, top 8 features)
selector = SelectKBest(score_func=f_classif, k=8)
X_selected = selector.fit_transform(X_resampled, y_resampled)
selected_features = X.columns[selector.get_support()]

print("Selected Features:", list(selected_features))

# 🧠 7. Train/Test Split
X_train, X_test, y_train, y_test = train_test_split(X_selected, y_resampled, test_size=0.2, random_state=42)

# 🌳 8. Train Classifier
clf = RandomForestClassifier(class_weight='balanced', random_state=42)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

# 📊 9. Results
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# 🔍 10. Feature Importance
importances = clf.feature_importances_
plt.figure(figsize=(8, 5))
plt.barh(selected_features, importances)
plt.xlabel("Feature Importance")
plt.title("Top Features Contributing to Emotion Classification")
plt.show()
```
